# OPyN tutorial

This tutorial will provide a brief demonstration of the functions implemented by the ``OPyN`` package and how they can be used to analyze complex time series.

To install ``OPyN``, run ``./install.sh`` in the OPyN directory.

```python
import numpy as np 
import matplotlib.pyplot as plt
import OPyN

# Some fMRI BOLD data
data = np.load("data.npz")["arr_0"].astype("float64") # OPyN is picky about data type. 

fig = plt.figure(dpi=150, figsize=(8,2))
ax = fig.add_subplot(1,1,1)
ax.plot(data)
ax.set_xlabel("Samples")
```

We begin by finding the optimal lag for the embedding - while there are many options, the most well-known is the first zero of the auto-correlation function, which is implemented here. 

This uses the Pearson autocorrelation, so it is only sensitive to linear temporal dependencies.

```python
# This code searches for the maximum lag uses the first zero of the autocorrelation function
lag_star = OPyN.optimal_lag(
    data,
    lrange=15,
    step=1)

print(lag_star)
```

Now we need to select the optimal embedding dimension. There are two options implemented. The first (and preferred optio) is to find the embedding dimension that maximizes the variance in the degree distribution of the resulting OPN. If the embedding dimension is too low, then every state will transition to every other state and the result will be a fully-connected network with low variance of degree. In contrast, if embedding dimension is too high, then each state will only appear once, creating a chain of points, all with in-degree and out-degree of one. 

A graph with high variance of the degree distribution is, in some sense, the "richest" graph, displaying both common, recurrent states and rarer states. 

```python
dim_star = OPyN.optimal_dim(
    data,
    lag = lag_star,
    drange=8)

print(dim_star)
```

We can also use a false-nearest neighbors method: this approach is based on finding the minimal number of dimensions required to correctly represent the nearest-neighbor structure (and by extension, the global topology) of the embedded point cloud. 

```python
dim_fnn = OPyN.false_nearest_neighbors(
    data,
    threshold = 0.05, #we want a threshold of only 5% remaining false nearest neighbors.
    lag = lag_star)

print(dim_fnn)
```

We can see that the variance-based embedding and the false-nearest-neighbors based embedding don't always match. This can be due to many reasons, including noise (the FNN approach is sensitive to noise while the variance approach has already been discretized), autocorrelation in the data, and the peculiarities of the distance function chosen.

We will use the ``dim_star`` for the purposes of this tutorial. 

Now we can construct the OPN

```python
opn, series = OPyN.ordinal_partition_network(
    data,
    dim_star,
    lag_star)
```

We can do some basic characterization of the OPN using ``igraph`` syntax:

```python
print(f"Number of vertices: {opn.vcount()}")
print(f"Number of edges: {opn.ecount()}")
```

Recall that the number of possible edges in ``dim_star!``

```python
num_possibilities = np.prod(np.arange(1, dim_star+1)[::-1])
print(f"Number of possible states: {num_possibilities}")
print(f"Proportion of possible states visited: {opn.vcount()/num_possibilities}")
print(" ")
print(f"Number of possible edges: {opn.ecount()**2}")
print(f"Proportion of possible edges visited: {opn.ecount() / opn.ecount()**2}")
```

This looks good: we get a decent sample of the total configuraiton space (but don't saturate it), but the network itself is quite sparse.

The ``ordinal_partition_network()`` function also returns the series itself, so we can get a look at the histogram of visits. 

```python
unq, counts = np.unique(series, return_counts=True)
argsort = np.argsort(counts)

fig = plt.figure(dpi=150)
ax = fig.add_subplot(1,1,1)

# ~700 nodes is too many to label each one, so we'll sort them by frequency (corresponding to in-degree).
ax.scatter(np.arange(1,1+unq.shape[0]),
           counts[argsort][::-1] / counts.sum(),
           s=1)
ax.set_xlabel("Frequency rank")
ax.set_ylabel("Proportion of time spent on vertex")
ax.set_yscale("log")
```

So we have somedting that looks reasonably heavy-tailed. There are some motifs that appear a lot, and others that appear much less, byt are still present. 

We can visualize the structure of the network by extracting its transition probability matrix:

```python
tpm, labels = OPyN.transition_probability_matrix(opn)
```

We can also compute some basic information-theoretic measures from the OPN.

* The entropy production measures how far from thermodynamic equilibrium the network is (i.e. how irreversible is it).
* The determinism quantifies how reliably the future can be predicted from the present.
* The degeneracy quantifies how much information is lost when past states "run together".
* The effective information quantifies how much information the system encodes in its causal structure.

```python
print(f"Entropy production: {OPyN.entropy_production(opn)} bit")
print(f"Determinism: {OPyN.determinism(opn)} bit")
print(f"Degeneracy: {OPyN.degeneracy(opn)} bit")
print(f"Effective information: {OPyN.effective_information(opn)} bit")
```

So we can see that the OPN from the fMRI BOLD time series is: very deterministic, weakly degenerate, and far from equilibirum.
